```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the sales data CSV file into a DataFrame
sales_df = pd.read_csv("sales_data.csv")

# Convert the 'Date' column to datetime format
sales_df['date'] = pd.to_datetime(sales_df['date'])

# Set the 'Date' column as the index
sales_df.set_index('date', inplace=True)

# Resample the data to monthly frequency and calculate the monthly totals
monthly_totals = sales_df.resample('M').sum()

# Plot the sales trends over time
plt.figure(figsize=(10, 6))
plt.plot(monthly_totals.index, monthly_totals['sales'], marker='o', linestyle='-')
plt.title('Sales Trends Over Time')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Group the sales data by product ID and calculate the total sales for each product
product_sales = sales_df.groupby('product_id')['sales'].sum()

# Sort the products by total sales in descending order
top_selling_products = product_sales.sort_values(ascending=False)

# Print the top-selling products
print("Top-Selling Products:")
print(top_selling_products.head())

```

```python
# Group the sales data by customer segment and calculate the total sales for each segment
segment_sales = sales_df.groupby('customer_segment')['sales'].sum()

# Sort the customer segments by total sales in descending order
top_customer_segments = segment_sales.sort_values(ascending=False)

# Print the top customer segments
print("\nTop Customer Segments:")
print(top_customer_segments)
```
